```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data from the system
```

```python
# reading the data
```

```python
df=pd.read_csv('1384439.csv')
df.head()
```

```python
# finding the runs scored in each inning
```

```python
columns_to_sum = ['Runs-off-bat', 'Extras', 'Wides', 'No-balls', 'Byes', 'Leg-byes', 'Penalty']
total_runs_per_inning = df.groupby('Innings number, starting from 1')[columns_to_sum].sum()
```

```python
total_runs_per_inning
```

```python
# making the visualization for the same
```

```python
plt.figure(figsize=(9, 5))
plt.subplot(1, 2, 1)
sns.barplot(total_runs_per_inning[total_runs_per_inning['Runs-off-bat']==228.0])
plt.title('inning 1')
plt.xticks(rotation=90)

# Create the second plot (e.g., bar plot)
plt.subplot(1, 2, 2)
sns.barplot(total_runs_per_inning[total_runs_per_inning['Runs-off-bat']==223.0])
plt.title('inning 2')
plt.xticks(rotation=90)
```

```python
# finding the total runs scored by each batsman
```

```python
sns.barplot(df.groupby('Batsman')['Runs-off-bat'].sum())
plt.xticks(rotation=90)
plt.show()
```

```python
wickets_df = df[df['Kind of wicket, if any'].notnull()]
wicket_df=wickets_df[['Innings number, starting from 1','Over and ball','Dismissed played, if there was a wicket']]
wicket_df
```

```python
wickets_df = df[(df['Innings number, starting from 1'] == 1) & df['Kind of wicket, if any'].notnull()]

# Convert 'Over and ball' to numeric for plotting
wickets_df['Over and ball'] = wickets_df['Over and ball'].apply(lambda x: float(x))

# Create a scatter plot for Inning 1
plt.figure(figsize=(10, 6))
plt.scatter(wickets_df['Over and ball'], [1] * len(wickets_df), label='Inning 1', color='red')

plt.xlabel('Over and Ball')
plt.ylabel('Inning')
plt.title('Fall of Wickets in Inning 1')
plt.legend()
plt.show()
```

```python
wickets_df = df[(df['Innings number, starting from 1'] == 2) & df['Kind of wicket, if any'].notnull()]

# Convert 'Over and ball' to numeric for plotting
wickets_df['Over and ball'] = wickets_df['Over and ball'].apply(lambda x: float(x))

# Create a scatter plot for Inning 1
plt.figure(figsize=(10, 6))
plt.scatter(wickets_df['Over and ball'], [1] * len(wickets_df), label='Inning 2', color='red')

plt.xlabel('Over and Ball')
plt.ylabel('Inning')
plt.title('Fall of Wickets in Inning 2')
plt.legend()
plt.show()
```

```python
wickets_df = df[df['Kind of wicket, if any'].notnull()]

# Count the number of wickets for each bowler
wickets_count = wickets_df['Bowler'].value_counts()

# Find the highest wicket-taker
highest_wicket_taker = wickets_count.idxmax()
highest_wickets = wickets_count.max()

print(f'Highest wicket-taker: {highest_wicket_taker} with {highest_wickets} wickets')
```
