# univarient

```python


def univariate_analysis(finals):
    plt.figure(figsize=(14, 8))
    sns.set(style="whitegrid")

    # Runs-off-bat distribution
    plt.subplot(2, 2, 1)
    sns.histplot(finals['Runs-off-bat'].dropna(), bins=10, kde=True, color="skyblue")
    plt.title('Distribution of Runs-off-bat')

    # Extras distribution
    plt.subplot(2, 2, 2)
    sns.histplot(finals['Extras'], bins=10, kde=True, color="salmon")
    plt.title('Distribution of Extras')

    # Wides distribution
    plt.subplot(2, 2, 3)
    sns.histplot(finals['Wides'].dropna(), bins=10, kde=True, color="lightgreen")
    plt.title('Distribution of Wides')

    # No-balls distribution
    plt.subplot(2, 2, 4)
    sns.histplot(finals['No-balls'].dropna(), bins=10, kde=True, color="orange")
    plt.title('Distribution of No-balls')

    plt.tight_layout()
    plt.show()
```

```python
univariate_analysis(semifinal1)
```

```python
univariate_analysis(semifinal2)
```

```python
univariate_analysis(finals)
```

# bivariate

```python
def bivariate_analysis(finals):
    plt.figure(figsize=(14, 6))

    # Runs-off-bat vs Extras
    plt.subplot(1, 2, 1)
    sns.scatterplot(x=finals['Runs-off-bat'], y=finals['Extras'], alpha=0.6, color="purple")
    plt.title('Runs-off-bat vs Extras')

    # Wides vs Extras
    plt.subplot(1, 2, 2)
    sns.scatterplot(x=finals['Wides'], y=finals['Extras'], alpha=0.6, color="teal")
    plt.title('Wides vs Extras')

    plt.tight_layout()
    plt.show()
```

```python
bivariate_analysis(semifinal1)
```

```python
bivariate_analysis(semifinal2)
```

```python
bivariate_analysis(finals)
```

# multivariate

```python
def multivariate_analysis(finals):
    plt.figure(figsize=(10, 6))

    # Selecting only numerical columns for correlation
    num_df = finals[['Runs-off-bat', 'Extras', 'Wides', 'No-balls']].dropna()

    # Plotting the correlation heatmap
    sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
multivariate_analysis(semifinal1)
```

```python
multivariate_analysis(semifinal2)
```

```python
multivariate_analysis(finals)
```

# match summary

```python
def match_summary(finals):
    # Aggregating runs, extras, wides, no-balls, and wicket count by 'Innings number'
    summary = finals.groupby("Innings number, starting from 1").agg({
        'Runs-off-bat': 'sum',
        'Extras': 'sum',
        'Wides': 'sum',
        'No-balls': 'sum',
        'Kind of wicket, if any': 'count'  # counting wickets
    }).rename(columns={'Kind of wicket, if any': 'Wickets'}).reset_index()

    print(summary)
    return summary
```

```python
match_summary(semifinal1)
```

```python
match_summary(semifinal2)
```

```python
match_summary(finals)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns


def player_clustering_analysis(semifinal1, players):

    data = semifinal2[['Runs-off-bat', 'Extras', 'Wides', 'No-balls']].fillna(0)


    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)


    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data_scaled)
        wcss.append(kmeans.inertia_)


    plt.figure(figsize=(8, 5))
    plt.plot(range(1, 11), wcss, marker='o', color='b')
    plt.title('Elbow Method for Optimal Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()


    kmeans = KMeans(n_clusters=3, random_state=42)
    semifinal1['Cluster'] = kmeans.fit_predict(data_scaled)


    player_data = semifinal1[semifinal1['Batsman'].isin(players)]
    non_player_data = semifinal1[~semifinal1['Batsman'].isin(players)]


    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=semifinal1['Runs-off-bat'], y=semifinal1['Extras'], hue=semifinal1['Cluster'], palette='viridis', s=100)
    sns.scatterplot(x=player_data['Runs-off-bat'], y=player_data['Extras'], color='red', s=200, label='Specific Players')
    plt.title('Clusters with Specific Players Highlighted')
    plt.xlabel('Runs-off-bat')
    plt.ylabel('Extras')
    plt.legend()
    plt.show()


    return player_data[['Batsman', 'Runs-off-bat', 'Extras', 'Wides', 'No-balls', 'Cluster']]
```
